```python
def left_rotate(source, k):
    fr = k
    for i in range(len(source)):
        print(source[fr], end="")
        fr-= 1
        if fr < 0:
            fr = len(source) - 1
    print()


def right_rotate(source, l):
    fr = l
    for i in range(len(source)):
        print(source[fr], end="")
        fr+= 1
        fr%= len(source)
    print()

def cap_finder(source):
    sc, si= None, None
    for i in range(len(source)):
        if 65<=ord(source[i])<=90:
            sc = source[i]
            si = i
    return sc, si

def billboard(s1, s2):
    mul_arr = [[""] * 10, [""] * 10]

    for i in range(len(s1)):
        mul_arr[0][i] = s1[i]

    for i in range(len(s2)):
        mul_arr[1][i] = s2[i]

    top_start_char, top_start_index= cap_finder(mul_arr[0])
    bottom_start_char, bottom_start_index= cap_finder(mul_arr[1])
    print(f"Top Board Start Character: {top_start_char}\nTop Board Start Index: {top_start_index}\nBottom Board Start: {bottom_start_char}\nBottom Board Start Index: {bottom_start_index}")

    k, l= top_start_index, bottom_start_index

    while 1:
        print("Press any key and then press enter to continue!!!")
        key = input()
        if key == "q" or key == "Q":
            tsc_checker, tsi_checker= cap_finder(mul_arr[0])
            bsc_checker, bsi_checker= cap_finder(mul_arr[1])
            print(f"Top Board Start Character: {tsc_checker}\nTop Board Start Index: {tsi_checker}\nBottom Board Start: {bsc_checker}\nBottom Board Start Index: {bsi_checker}")
            break
        else:
            left_rotate(mul_arr[0], k)
            right_rotate(mul_arr[1], l)
            k-= 1
            if k<0:
                k= len(mul_arr[0])-1
            l-= 1
            if l<0:
                l= len(mul_arr[1])-1



string_1 = input()
if len(string_1) > 10:
    print("Invalid Input Size")
else:
    string_2 = input()
    if len(string_2) > 10:
        print("Invalid Input Size")
    else:
        print("Your input is valid. Press any key and then press enter to continue!!!")
        billboard(string_1,string_2)
```

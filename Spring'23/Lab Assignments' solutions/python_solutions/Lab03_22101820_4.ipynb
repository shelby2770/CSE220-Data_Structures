```python
# Run This cell Second
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
# To you coding in this cell
class LinkedList:

  def __init__(self, a):
    if type(a)== list:
        self.head= Node(a[0],None)
        shelby= self.head
        for i in range(1, len(a)):
          new_node= Node(a[i], None)
          shelby.next, shelby= new_node, new_node  
    else:
      self.head= a


  # Traverse elements in the list.
  # This method is done for you. Do not change this method.
  def traverseList(self):
    s = ''
    temp = self.head
    while temp != None:
      if temp.next != None:
        s += str(temp.element) + " "
      else:
        s += str(temp.element)
      temp = temp.next
    return s
  

  # Count the number of nodes in the list and return the total number
  def countNode(self):
    temp= self.head
    counter= 0
    while temp!= None:
      temp= temp.next
      counter+= 1
    return counter
  

  # returns the reference of the Node at the given index. For invalid index return None.
  def nodeAt(self, idx):
    length= self.countNode()
    temp= self.head
    if idx<0 or idx>= length:
      return None
    else:
      for i in range(length):
        if i== idx:
          return temp
        temp= temp.next

  
  # returns the element of the Node at the given index. For invalid idx return None.
  def get(self, idx):
    return self.nodeAt(idx).element
  
  
  # updates the element of the Node at the given index. 
  # Returns the old element that was replaced. For invalid index return None.
  # parameter: index, element
  def set(self, idx, elem):
    length= self.countNode()
    val= self.head.element
    if idx<0 or idx>= length:
      return None

    if idx== 0:
      self.head.next, self.head= None, Node(elem, self.head.next)
    else:
      val= self.nodeAt(idx).element
      new_node= Node(elem, self.nodeAt(idx).next)
      self.nodeAt(idx).next, self.nodeAt(idx-1).next= None, new_node
    return val
   

  # returns the index of the Node containing the given element.
  # if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    length= self.countNode()
    temp= self.head
    idx= 0

    while temp.element!= elem:
      temp= temp.next
      idx+= 1
      if idx== length:
        return -1
    return idx
  
  
  # returns true if the element exists in the List, return false otherwise.
  def contains(self, elem):
    return self.indexOf(elem)!= -1


  # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
  def copyList(self):
    length= self.countNode()
    new_list= [None]*length
    temp= self.head
    counter= 0
    while temp.next!= None:
      new_list[counter]= temp.element
      counter+= 1
      temp= temp.next
    new_list[length-1]= temp.element
    return new_list


  # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
  def reverseList(self):
    length= self.countNode()
    new_list= [None]*length
    temp= self.head
    counter= length-1
    while temp.next!= None:
      new_list[counter]= temp.element
      counter-= 1
      temp= temp.next
    new_list[0]= temp.element
    return new_list


  # inserts Node containing the given element at the given index
  # Check validity of index. If invalid then print "Invalid Index"
  def insert(self, elem, idx):
    length= self.countNode()
    if idx<0 or idx>length:
      print("Invalid Index")
    else:
      if idx== 0:
        self.head= Node(elem, self.head)
      else:
        self.nodeAt(idx-1).next= Node(elem, self.nodeAt(idx))

    
  # removes Node at the given index. returns element of the removed node.
  # Check validity of index. return None if index is invalid.
  def remove(self, idx):
    length= self.countNode()
    rem_val= self.head.element
    if idx<0 or idx>= length:
      return None
      
    if idx== 0:
      self.head, self.head= None, self.head.next
    else:
      rem_val= self.nodeAt(idx).element
      self.nodeAt(idx).next, self.nodeAt(idx-1).next= None, self.nodeAt(idx).next       
    return rem_val
          
  
  # Rotates the list to the left by 1 position.
  def rotateLeft(self):
    length= self.countNode()
    if length>1:
      temp= self.head.element
      self.head.next, self.head= None, self.head.next
      self.nodeAt(length-2).next= Node(temp, None)
    

  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    length= self.countNode()
    if length>1:
      temp1= self.nodeAt(length-2)
      temp2= temp1.next
      temp1.next= None
      temp2.next= self.head
      self.head= temp2
```

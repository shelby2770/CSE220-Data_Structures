```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
    self.head= Node(a[0],None, None)
    shelby= self.head
    for i in range(1, len(a)):
      new_node= Node(a[i], None, None)
      temp= shelby
      shelby.next, shelby= new_node, new_node  
      shelby.prev= temp
    shelby.next= self.head
    self.head.prev= shelby


  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    temp= self.head
    counter= 1
    while temp.next!= self.head:
      temp= temp.next
      counter+= 1
    return counter
  

  # prints the elements in the list
  def forwardprint(self):
    temp= self.head
    while temp.next!= self.head:
      print(temp.element, end= ",")
      temp= temp.next
    print(temp.element)
    

  # prints the elements in the list backward
  def backwardprint(self):
    length= self.countNode()
    temp= self.nodeAt(length-1)
    while temp!= self.head:
      print(temp.element, end= ",")
      temp= temp.prev
    print(temp.element)


  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    length= self.countNode()
    temp= self.head
    if idx<0 or idx>= length:
      new_node= Node("invalid idx", None, None)
      return new_node #if I return None, then for any invalid index, myNode.element will show an error.
    else:
      for i in range(length):
        if i== idx:
          return temp
        temp= temp.next
  

  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    length= self.countNode()
    temp= self.head
    idx= 0
    while temp.element!= elem:
      temp= temp.next
      idx+= 1
      if idx== length:
        return -1
    return idx


  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    length= self.countNode()
    if idx<0 or idx>length:
      print('Invalid index')
    else:
      new_node= Node(elem, None, None)
      temp_1= self.nodeAt(idx-1) if idx!= 0 else self.nodeAt(length-1)
      temp_2= temp_1.next
      temp_1.next, new_node.prev= new_node, temp_1
      temp_2.prev, new_node.next= new_node, temp_2
      if idx== 0:
        self.head= new_node


  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    length= self.countNode()
    if idx<0 or idx>=length:
      return "None"
    
    rem_node= self.nodeAt(idx)
    rem_val= rem_node.element
    temp_1= rem_node.prev
    temp_2= rem_node.next
    rem_node.prev, rem_node= None, None
    temp_1.next, temp_2.prev= temp_2, temp_1
    if idx== 0:
        self.head= temp_2
        
    return str(rem_val)
```

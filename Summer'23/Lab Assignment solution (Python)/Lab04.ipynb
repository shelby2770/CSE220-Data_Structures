```python
class Patient:
  def __init__(self, id, name, age, bloodgroup, next= None, prev= None):
    self.id= id
    self.name= name
    self.age= age
    self.bloodgroup= bloodgroup
    self.next = next # To store the next node’s reference.
    self.prev = prev # To store the previous node’s reference.

class WRM:
  def __init__(self):
    self.head= Patient(-1, '', -1, '') #dummy head
    self.head.next= self.head
    self.head.prev= self.head

  def RegisterPatient(self, id, name, age, bloodgroup):
    prev_patient= self.head.prev
    new_patient= Patient(id, name, age, bloodgroup)
    prev_patient.next= new_patient
    new_patient.prev= prev_patient
    self.head.prev= new_patient
    new_patient.next= self.head
    print("Success registering patient")

  def ServePatient(self):
    if self.head.next== self.head:
      print("No Patient to be served")
      return

    print(f"{self.head.next.name} is served")
    self.head.next= self.head.next.next
    self.head.next.prev= self.head

  def CancelAll(self):
    self.head.next= self.head
    self.head.prev= self.head
    print("Success")

  def CanDoctorGoHome(self):
    print("No" if self.head.next!= self.head else "Yes")

  def ShowAllPatient(self):
    if self.head.next== self.head:
      print("No patient in wrm")
      return

    k= self.head.next
    while k!= self.head:
      print(k.name, end= " ")
      k= k.next
    print()


wrm_system= WRM()
print("Welcome to Waiting Room Management System")


while 1:
  print("**Choose an option**\n1. RegisterPatient()\n2. ServePatient()\n\
3. CancelAll()\n4. CanDoctorGoHome():\n5. ShowAllPatient()")
  n= int(input("Enter your choice: "))
  if n== 1:
    print("Executing RegisterPatient()...")
    id= int(input("Enter ID: "))
    name= input("Enter Name: ")
    age= int(input("Enter Age: "))
    bloodgroup= input("Enter Blood Group: ")
    wrm_system.RegisterPatient(id, name, age, bloodgroup)

  elif n== 2:
    print("Executing ServePatient()...")
    wrm_system.ServePatient()

  elif n== 3:
    print("Executing CancelAll()...")
    wrm_system.CancelAll()

  elif n== 4:
    print("Executing CanDoctorGoHome()...")
    wrm_system.CanDoctorGoHome()

  elif n== 5:
    print("Executing ShowAllPatient()...")
    wrm_system.ShowAllPatient()

  elif n== 6:
    print("Thank You For Using Waiting Room Management System\nEXITING...")
    break

  else:
    print("No such option")
  print()
```
